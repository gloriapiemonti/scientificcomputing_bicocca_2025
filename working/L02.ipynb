Q2: Histograms
-

```python
url = "https://raw.githubusercontent.com/sbu-python-summer/python-tutorial/master/day-3/sample.txt"
filename = "sample.txt"
r = requests.get(url, allow_redirects=True)

with open(filename, "w") as f:     
    f.write(r.text) 

a = np.loadtxt(filename)
counts, bin_edges = np.histogram(a)

for i in range(len(counts)):
    bin_center = (bin_edges[i] + bin_edges[i+1])/2
    print("Bin center: ", bin_center, "Counts: ", counts[i])
```

Q3: Are you faster than numpy? No
-

```python
a = np.linspace(1, 100, 10, endpoint=True)

t0 = time.time()
average = a.sum()/10
sum_square = 0

for i in a:
    b = (i - average)**2
    sum_square += b 
    
std = np.sqrt(sum_square/10)
t1 = time.time()
print("My function took: {} s".format(t1-t0))

t2 = time.time()
std_dev = np.std(a)
t3 = time.time()
print("Numpy took: {} s".format(t3-t2))

print("My function:", std, ", Numpy:", std_dev)
```
