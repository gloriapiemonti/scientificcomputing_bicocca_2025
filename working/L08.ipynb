```python
#Random walk simulation

def random_walk_2d(steps):
    x, y = 0, 0
    position = [(x, y)]

    for i in range(steps):
        dx, dy = random.choice([(0, 1), (0, -1), (1, 0), (-1, 0)])
        x += dx
        y += dy
        position.append((x, y))

    return position


steps = 10**8
position = random_walk_2d(steps)
print("Final Position:", position[-1])
```

```python
cProfile.run('random_walk_2d(10**6)', sort='cumtime')
```

Main bottleneck: random.choice with cumtime = 0.86

```python
@njit
def optimized_random_walk_2d(steps):
    choices = np.array([-1, 1])
    position = np.zeros(2)

    for i in range(steps):
        axis = np.random.randint(0, 2)
        position[axis] += choices[np.random.randint(0, 2)]
    return position


steps = 10**8
final_pos = optimized_random_walk_2d(steps)
print("Final Position:", final_pos)
```

```python
optimized_random_walk_2d(1)          # warm-up JIT

start = time.perf_counter()
optimized_random_walk_2d(10**8)
print(f"Numba: {time.perf_counter() - start:.3f}s")
```
