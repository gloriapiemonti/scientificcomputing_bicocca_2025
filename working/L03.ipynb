Q2: Planetary positions
-

```python
a = np.array([0.39, 0.72, 1.00, 1.52, 5.20, 9.54, 19.22, 30.06, 39.48])
P = np.array([0.24, 0.62, 1.00, 1.88, 11.86, 29.46, 84.01, 164.8, 248.09])
names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", 
         "Uranus", "Neptune", "Pluto"]

plt.plot(a, P, "o")
plt.xscale("log")
plt.yscale("log")
plt.xlabel(r"$Distance [AU]$")
plt.ylabel(r"Period [years]")
for i in range(9):
    plt.text(a[i], P[i], names[i])
```

Q3: A perfect circle
-

```python
N = 50
R = 20
theta = np.linspace(0, 2*np.pi, N)

fig = plt.figure()

ax1 = fig.add_subplot(211)
ax1.plot(R*np.cos(theta), R*np.sin(theta), "o", color="darkviolet")
ax1.axis("equal")

ax2 = fig.add_subplot(212)
ax2.fill(R*np.cos(theta), R*np.sin(theta), facecolor='darkorchid', edgecolor='darkviolet', linewidth=3)
ax2.axis("equal")
```

```python
def draw_circle(ax, x0, y0, R, color):
    theta = np.linspace(0, 2*np.pi, 200)
    ax.plot(x0 + R * np.cos(theta), y0 + R * np.sin(theta), "o", color=color)
    ax.axis("equal")


fig, ax = plt.subplots()
colors = {1: "royalblue", 2: "crimson", 3: "magenta", 4: "lightseagreen", 5: "darkorchid"}

for i in range(10):
    x0 = random.randint(1, 20)
    y0 = random.randint(1, 20)
    R = random.randint(1, 10)
    color = colors[random.randint(1, 5)]
    draw_circle(ax, x0, y0, R, color)

plt.show()
```

Q6: frequent words plotting
-

```python
f = open("../lectures/shore_leave.txt", "r")
count = {}
text = ""
for line in f:
    text += line 
w = text.split()
words = [q.lower().replace(".", "").replace(",", "") for q in w]
unique_words = set(words)

for i in unique_words:
    count[i] = words.count(i)

fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(211)

sorted_words = sorted(count.items(), key=lambda x: x[1], reverse=True)
top_words = [w for w, c in sorted_words[:25]]
top_counts = [c for w, c in sorted_words[:25]]

bar_container = ax.bar(top_words, top_counts)
ax.set(
    ylabel="Counts",
    title="25 most frequent words",
)
ax.bar_label(bar_container, fmt='{:,.0f}')
```
